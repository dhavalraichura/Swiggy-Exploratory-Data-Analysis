```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv("../../Datasets/Swiggy/swiggy_file.csv")
df
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
df['Number of Ratings'] = df['Number of Ratings'].str.replace('+ ','')
df['Number of Ratings'] = df['Number of Ratings'].str.replace('ratings','')
df['Number of Ratings'] = df['Number of Ratings'].str.replace('Too Few Ratings','0')
df['Number of Ratings'] = df['Number of Ratings'].str.replace('K','000')
df
```

```python
df['Rating'] = df['Rating'].str.replace('--','0')
df
```

```python
df['Average Price'] = df['Average Price'].str.replace(' for two','')
df['Average Price'] = df['Average Price'].str.replace('₹','')
df['Average Price'] = (df['Average Price'].astype(int))
df['Average Price'] = df['Average Price'] / 2
df
```

```python
df['Offer Name'] = df['Offer Name'].str.replace('\n',',')
df
```
